```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/Fraud detection Python/creditcard.csv')
```

```python
print(dataset)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset['Class'].value_counts()
```

Highly unbalanced dataset as more than 90 % of the dataset contains legit transactions

0 -> Legit

1 -> Fraud

```python
legit = dataset[dataset.Class == 0]
fraud = dataset[dataset.Class == 1]
```

```python
print(legit)
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
print(legit.Amount.describe())
print(fraud.Amount.describe())
```

```python
dataset.groupby('Class').mean()
```

Under sampling

Taking a sample of legit transactions 

```python
legit_sample=legit.sample(n=492)
```

```python
print(legit_sample)
```

Concatenating Legit_sample with fraud transactions

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head(10)
```

Selecting Target variable and Training variables

```python
x=new_dataset.iloc[:,:-1]
```

```python
x.head()
```

```python
y=new_dataset.iloc[:,-1]
```

Dividing the variables in training and testing data

Training -> 70%

Testing -> 30%

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3,stratify=y)
```

```python
x_train.head()
```

```python
x_test.head()
```

Now Finally training our model on training data

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

Creating Confusion Matrix

```python
conf_matrix=confusion_matrix(y_true=y_test, y_pred=y_pred)
print(conf_matrix)
```

Calculating Precision and Recall

```python
precision=conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[1][0])
print(precision," %")
```

```python
recall=conf_matrix[0][0]/(conf_matrix[0][0]+conf_matrix[0][1])
print(recall," %")
```

Calculating Accuracy of the model

```python
print(accuracy_score(y_pred,y_test))
```
